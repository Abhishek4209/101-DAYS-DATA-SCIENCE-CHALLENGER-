```python
df["Date"]=df["Date_of_Journey"].str.split("/").str[0].astype("int64")
df["Month"]=df["Date_of_Journey"].str.split("/").str[1].astype("int64")
df["Year"]=df["Date_of_Journey"].str.split("/").str[2].astype("int64")
```

```python
df.drop(["Date_of_Journey"],axis=1,inplace=True)
```

```python
df["Arrival_Time"].str.split(" ").str[0]
```

```python
df["Arrival_hours"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[0].astype("int64")
df["Arrival_minutes"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[1].astype("int64")
df.drop(["Arrival_Time"],axis=1,inplace=True)
```

```python
df["Dep_hour"]=df["Dep_Time"].str.split(":").str[0].astype("int64")
df["Dep_minute"]=df["Dep_Time"].str.split(":").str[1].astype("int64")
df.drop(["Dep_Time"],axis=1,inplace=True)
```
