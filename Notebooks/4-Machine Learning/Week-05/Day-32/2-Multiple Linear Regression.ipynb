````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
````

````python
df=sns.load_dataset("tips")
````

````python
df.head()
````

````python
df.isnull().sum()
````

````python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelBinarizer
encoder1=LabelEncoder()
encoder2=LabelBinarizer()
````

````python
df["sex"]=encoder2.fit_transform(df["sex"])
df["smoker"]=encoder2.fit_transform(df["smoker"])
df["time"]=encoder2.fit_transform(df["time"])
df["day"]=encoder1.fit_transform(df["day"])
df.head()
````

````python
df.head()
````

````python
x=df.drop(["total_bill"],axis=1)
y=df["total_bill"]
````

````python
sns.scatterplot(x=df["total_bill"],y=df["tip"],hue=df["sex"])
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
````

````python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
````

````python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
````

````python
model.fit(x_train,y_train)
````

````python
y_pred=model.predict(x_test)
y_pred
````

````python
from sklearn.metrics import r2_score
````

````python
r2_score(y_test,y_pred)
````
